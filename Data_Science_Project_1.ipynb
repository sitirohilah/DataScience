21. Bagaimana cara meringkas jumlah unik customer, jumlah unik invoice number, jumlah quantity, jumlah revenue, jumlah unik deskripsi, mengurutkan price dari terkecil ke terbesar, dan mengurutkan invoice price dari terbesar ke terkecil? Gunakan `def ... condition`

```python
def ringkas_data(df):
  return(
      df.CustomerID.unique(),
      df.InvoiceNo.unique(),
      df.Quantity.sum(),
      df.Revenue.sum(),
      df.Description.nunique(),
      df.sort_values(by='UnitPrice', ascending=False).head(),
      df.sort_values(by='UnitPrice').head(),
  )
ringkas_data(df)
```

# Prepare Dataset Untuk RFM

27. Bagaimana cara mengecek tanggal invoice terbesar?

```python
today = df.InvoiceDate.max()
```

```python
df.describe()
```

28. Buatlah recency, frequency, dan monetary pada dataset online retail!

*   recency menunjukkan waktu interaksi terakhir pelanggan dengan produk.
*   Frequency mengacu pada berapa kali pelanggan berinteraksi dengan produk atau melakukan transaksi dalam periode waktu tertentu
* Monetary value menunjukkan jumlah total yang dihabiskan oleh pelanggan untuk membeli produk Anda dalam periode waktu tertentu.

```python
data_rfm=df.groupby('CustomerID').agg({
    'InvoiceDate' : lambda date: (today-date.max()).days,
    'InvoiceNo' : 'count',
    'UnitPrice' : 'sum'
})

data_rfm.head()
```

29. ubahlah nama kolom pada data_rfm menjadi recency,frequency, dan monetory

```python
data_rfm.columns = ['Recency','Frequency','Monetory']
data_rfm.head()
```

31. Bagaimana cara extract tahun, bulan, hari, dan jam?

```python
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['Day'] = df['InvoiceDate'].dt.day
df['Hour'] = df['InvoiceDate'].dt.hour

df.head()
```

32. Hitunglah data berdasarkan InvoiceYearMonth!

```python
df['InvoiceYearMonth'] = df['InvoiceDate'].dt.to_period('M')

df.groupby('InvoiceYearMonth').agg({
    'InvoiceNo' : 'count',
    'UnitPrice' : 'sum'
}).head()
```

33. Bagaimana cara mengklasifikasi Quantity <= 6 True dan lainnya False?

Applying IF condition with Lambda

```python
df['classified'] = df['Quantity'].apply(lambda x: True if x>=6 else False)
df.head()
```

34. Bagaimana cara mengklasifikasi Quantity >= 6 adalah angka Quantity dan lainnya menjadi Below Threshold?

Applying IF condition Using pd.where Method

```python
df['classified'] = df['Quantity'].where(df['Quantity'] >= 6, other='Below Threshold')
df.tail()
```

39. Bagaimana pertumbuhan quantity dan revenue dari tahun ke tahun pada dataset nomor 38?

```python
yearly_data = df.groupby('Year').agg({'Quantity': 'sum', 'Revenue': 'sum'})
yearly_data['Quantity Growth'] = yearly_data['Quantity'].pct_change() * 100
yearly_data['Revenue Growth'] = yearly_data['Revenue'].pct_change() * 100
yearly_data
```

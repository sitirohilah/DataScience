**4. Ada berapa transaksi gratis (0 rupiah) yang terdapat dalam data? Siapakah penumpang yang melakukan transaksi perjalanan dengan yang melewati jumlah pemberhentian terbanyak (stopEndSeq - stopStartSeq) dan sebutkan rute perjalanan penumpang tersebut?**

```python
df_add['num_stop'] = df_add['stopEndSeq'] - df_add['stopStartSeq']
gratis = df_add.query('payAmount==0')
total = df_add.sort_values('num_stop', ascending=False, ignore_index=True)
print("Jumlah transaksi gratis = ", len(gratis))
print(f"Penumpang dengan pemberhentian terbanyak : {df_add.loc[0,'payCardName']} rute {df_add.loc[0, 'tapInStopsName']} - {df_add.loc[0, 'tapOutStopsName']}")
```

**5. Berapakah biaya yang dikeluarkan dan rute perjalanan untuk transaksi dengan jarak tempuh terlama dan jumlah pemberhentian terbanyak?**

```python
df_add1 = df.copy()
df_add1['tapInTime'] = pd.to_datetime(df_add1['tapInTime'])
df_add1['tapOutTime'] = pd.to_datetime(df_add1['tapOutTime'])
df_add1['num_stop'] = df_add1['stopEndSeq'] - df_add1['stopStartSeq']
df_add1['duration'] = (df_add1['tapOutTime'] - df_add1['tapInTime']).dt.total_seconds() / 60
df_add1.head()
```

```python
df_add1 = df_add1.sort_values(by=['duration','num_stop'], ascending=False, ignore_index=True)
print(f"rute {df_add1.loc[0, 'tapInStopsName']} - {df_add1.loc[0, 'tapOutStopsName']} dengan biaya : {df_add1.loc[0,'payAmount']}")
```
